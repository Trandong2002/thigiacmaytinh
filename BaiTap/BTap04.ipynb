```python
# import thư viện
from PIL import Image
from imgtools import*
```

```python
# Đường dẫn ảnh 
my_path = 'C:/Users/HP/Hinhanh/05.JPG'
```

```python
# Đọc ảnh
img = load_image(my_path)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (90,70,170,170)
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán 
# Xác định vị trí dán
paste_position = (10, 50)
img.paste(cropped_image, paste_position)
display(img)
```
