```python
# Thêm thư viện
from PIL import Image
```

```python
# Khai báo một biến chứa thư mục ( / hoặc \\)
dir = 'C:/Users/HP/Hinhanh'
```

```python
# Khai báo đường dẫn ảnh
image_path = dir + '/01.JPG'
print(image_path)
```

```python
# Sử dụng Image.open() để đọc ảnh
HPHinhanh = Image.open(image_path)
```

```python
# Hiển thị ảnh
display(HPHinhanh)
```

```python
# In thông tin cơ bản 
print(" Định dạng ảnh: ", HPHinhanh.format)
```

```python
# In kích cỡ ảnh
print("Kích cỡ ảnh: ", HPHinhanh.size)
```

```python
#Lưu ảnh và chuyển đổi ảnh
new_HP_Hinhanh_path = dir + "/new_01.JPG"
HPHinhanh.save(new_HP_Hinhanh_path)
```

```python
# Đóng tập tin
HPHinhanh.close()
```
