```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đường dẫn ảnh
my_path = 'C:/Users/HP/Hinhanh/04.JPG' 
# Đọc ảnh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# Chuyển sang ảnh xám
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(), bins=250)
plt.title('Biểu đồ histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
# Làm nổi bậc đường biên
plt.figure()
plt.gray()
plt.contour(im_gray, origin='image')
plt.show()
```

```python
# Đường Dẫn Ảnh
my_path = 'C:/Users/HP/Hinhanh/04.JPG'

# Đọc ảnh và hiển thị ảnh với trục tọa độ
im = Image.open(my_path).convert('L')

# Hiển thị ảnh trên trục tọa độ
plt.imshow(im)

# Tạo hình vẽ mới
plt.figure()

# Đặt màu sắc là màu xám
plt.gray()

# Vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(im, origin="image")

# Đảm bảo tỷ lệ giữa trục x và trục y
plt.axis('equal')

# Hiển thị ảnh
plt.show()
```
